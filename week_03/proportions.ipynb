```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Popular_Baby_Names.csv')
df
```

```python
year_of_birth = df['Year of Birth'].value_counts()

plt.pie(year_of_birth, labels=year_of_birth.index, autopct='%.1f %%')
plt.title('Year of Birth')

plt.show()
```

```python
gender = df['Gender'].value_counts()
plt.pie(gender, labels=gender.index, autopct='%.1f%%')
plt.title('Gender')

plt.show()
```

```python
plt.pie(year_of_birth, labels=year_of_birth.index, autopct='%.1f%%', wedgeprops=dict(width=0.3))
plt.title('Year of Birth')

plt.show()
```
