```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('birds.csv')
df.head()
```

```python
for i in range(len(df)):
    if df['Name'][i] == 'Snow goose':
        plt.bar(df['Name'][i], df['MaxLength'][i], label='MaxLength')
        plt.bar(df['Name'][i], df['MinLength'][i], label='MinLength')
        plt.title('Length')

plt.legend()
plt.show()
```

```python
plt.ylabel('MaxBodyMass')
for i in range(len(df)):
    if df['Name'][i] == 'Snow goose':
        snow_geese = df['Name'][i]
        max_body = df['MaxBodyMass'][i]
        plt.bar(snow_geese, max_body, color='g')

    if df['Name'][i] == "Ross's goose":
        ross_geese = df['Name'][i]
        max_body = df['MaxBodyMass'][i]
        plt.bar(ross_geese, max_body, color='b')

plt.show()
```

```python
plt.rcParams['figure.figsize'] = [6, 6]
plt.tick_params(labelbottom=False,bottom=False)
plt.xlabel('birds')
plt.ylabel('maxwingspan')
plt.title('Max Wingspan in centimeter')
for i in range(len(df)):
    x = df['Name'][i]
    y = df['MaxWingspan'][i]
    plt.scatter(x, y)

    if df['Name'][i] == 'Snow goose':
        plt.text(x, y, df['Name'][i])

plt.show()        
```

```python
plt.tick_params(labelbottom=False,bottom=False)
plt.xlabel('birds')
plt.ylabel('maxwingspan')
plt.title('Max Wingspan in centimeter')
for i in range(len(df)):
    x = df['Name'][i]
    y = df['MaxWingspan'][i]

    if df['MaxWingspan'][i] < 500:
        plt.scatter(x, y)

    if df['Name'][i] == 'Snow goose':
        plt.text(x, y, df['Name'][i])
            
plt.show()
```

```python
Ducks_Geese_Waterfowl = df[df['Category'] == 'Ducks/Geese/Waterfowl']
Ducks_Geese_Waterfowl.plot(x='Name', kind='barh', title='Ducks/Geese/Waterfowl', stacked=True)

plt.show()
```
