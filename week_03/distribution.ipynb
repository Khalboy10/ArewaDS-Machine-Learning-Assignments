```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Popular_Baby_Names.csv')
df.head()
```

```python
df.plot(kind='hist', bins=40, figsize=(10,8))
```

```python
df['Year of Birth'].plot(kind='hist', bins=20, figsize=(10,8))

plt.show()
```

```python
df['Rank'].plot(kind='hist', bins=20, figsize=(10,8))

plt.show()
```

```python
filtered = df[(df['Year of Birth'] == 2015) & (df['Rank'] < 40)]
filtered.plot(kind='hist', bins=20, figsize=(10,8))

plt.show()
```

```python
filtered['Count'].plot(kind='hist', bins=40, figsize=(10,8))

plt.show()
```
