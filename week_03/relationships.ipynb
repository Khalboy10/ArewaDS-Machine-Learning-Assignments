```python
import pandas as pd
import matplotlib.pylab as plt
import seaborn as sns
```

```python
honey = pd.read_csv('Honey.csv')
honey.head()
```

```python
sns.relplot(data=honey, x='year', y='totalprod', hue='state')
plt.title('Honey Production Over the Years')
plt.xlabel('Year')
plt.ylabel('Total Production')

plt.show()
```

```python
sns.relplot(data=honey, x='year', y='totalprod', hue='state', kind='line')
plt.title('Honey Production Over the Years')
plt.xlabel('Year')
plt.ylabel('Total Production')

plt.show()
```

```python
sns.relplot(x='state', y='totalprod', data=honey, kind='line', height=10, aspect=2)
plt.title('Total Production of Each state')

plt.show()
```

```python
sns.relplot(
    data=honey, 
    x="year", y="totalprod",
    col="state", 
    col_wrap=3,
    kind="line")
```
