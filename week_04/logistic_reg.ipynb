```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
pumpkins = pd.read_csv('US_pumpkins.csv')
pumpkins.head()
```

```python
pumpkins.isnull().sum()
```

```python
df = {
    'City Name' : pumpkins['City Name'],
    'Package' : pumpkins['Package'],
    'Variety' : pumpkins['Variety'],
    'Low Price' : pumpkins['Low Price'],
    'High Price' : pumpkins['High Price'],
    'Date' : pumpkins['Date'],
    'Color' : pumpkins['Color'],
    'Origin' : pumpkins['Origin'],
    'Item Size' : pumpkins['Item Size'],
    'Repack' : pumpkins['Repack']
}
new_pumpkins = pd.DataFrame(df)
new_pumpkins
```

```python
new_pumpkins = new_pumpkins.apply(LabelEncoder().fit_transform)
new_pumpkins.info
```

```python
from sklearn.model_selection import train_test_split

features = ['City Name', 'Package', 'Variety', 'Low Price', 'High Price', 'Date', 'Origin', 'Item Size', 'Repack']
target = 'Color'

X = new_pumpkins[features]
y = new_pumpkins[target]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.23, random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=2000, solver='sag', penalty='l2')
model.fit(X_train, y_train)
prediction = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report

print(classification_report(y_test, prediction))
print('Predicted labels: ', prediction)
print('Accuracy: ', accuracy_score(y_test, prediction))
```
